```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'], axis=1)
#Irrelavant data
```

```python
data
```

```python
label_encoder_gender = LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
# Initialize One-Hot Encoder
onehot_encoder_geo = OneHotEncoder(sparse_output=False)

# Fit and transform the Geography column
geo_encoded = onehot_encoder_geo.fit_transform(data[['Geography']])

# Convert to DataFrame
geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=onehot_encoder_geo.get_feature_names_out(['Geography']),
    index=data.index
)

# Drop original Geography column and concatenate encoded columns
data = pd.concat([data.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
data.head()
```

```python
X = data.drop('Exited', axis = 1)
y = data['Exited']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.250,random_state=42)
```

```python
scaler= StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
with open('standard_scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
data.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)), ##HL1 connect to input layer
    Dense(32,activation='relu'), #HL2
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
